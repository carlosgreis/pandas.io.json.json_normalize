```python
#libraries
import json
import pandas as pd

#normalize method
from pandas.io.json import json_normalize as jnorm

#display size
pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 1000)
```

```python
pd.set_option('display.max_rows', 4000)
```

```python
#open json file as python list
with open('Totvs/dataset.txt') as dataFile:
    dataset = json.load(dataFile)
```

```python
#first try to flatten list: parse all
df0 = pd.DataFrame(dataset)
df0.head()
```

```python
#second try to flatten list: parse all
df1 = jnorm(dataset)
df1.head()
```

```python
#third try to flatten list: parse 'dets'
df2 = jnorm(dataset, 'dets')
df2.head()
```

```python
#third try to flatten list: parse 'prod', inside Â´dets'
df3 = jnorm(dataset, record_path = ['dets','nItem'], meta = [['dets','prod','indTot'],['dets','prod','qCom'],['dets','prod','uCom'],['dets','prod','vProd'],['dets','prod','vUnCom'],['dets','prod','xProd']], errors='ignore')
df3.head()
```
